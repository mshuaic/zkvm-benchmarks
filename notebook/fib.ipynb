```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datadir = "/home/mark/zkvm-benchmarks/benchmark_outputs/"
jolt=pd.read_csv(datadir + "fiboancci_jolt.csv")
risczero=pd.read_csv(datadir + "fibonacci_risczero.csv")
sp1=pd.read_csv(datadir + "fibonacci_sp1.csv")
```

```python
df = pd.DataFrame({"jolt":jolt["prover time (ms)"], "risczero":risczero["prover time (ms)"], "sp1":sp1["prover time (ms)"], "n":jolt["n"]})
ax= df.plot.barh(x="n")
for container in ax.containers:
    ax.bar_label(container, labels=[f'{e:.2f}x' for e in container.datavalues / df["jolt"]])
plt.xlabel('prover time (ms)')
plt.ylabel('n')
plt.title('Prover time for fibonacci(n) in different zkVMs')
plt.savefig("fib_time.png")
```

```python
df = pd.DataFrame({"jolt":jolt["proof size (bytes)"], "risczero":risczero["proof size (bytes)"], "sp1":sp1["proof size (bytes)"], "n":jolt["n"]})
ax= df.plot.barh(x="n")
for container in ax.containers:
    ax.bar_label(container, labels=[f'{e:.2f}x' for e in container.datavalues / df["jolt"]])
plt.xlabel('proof size (bytes)')
plt.ylabel('n')
plt.title('Proof size for fibonacci(n) in different zkVMs')
plt.savefig("fib_size.png")
```
