```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datadir = "/home/mark/zkvm-benchmarks/benchmark_outputs/"
jolt=pd.read_csv(datadir + "sha2_chain_jolt.csv")
risczero=pd.read_csv(datadir + "sha2_chain_risczero.csv")
sp1_1=pd.read_csv(datadir + "sha2_chain_sp1_1_shard.csv")
sp1_4=pd.read_csv(datadir + "sha2_chain_sp1_4_shard.csv")
sp1_16=pd.read_csv(datadir + "sha2_chain_sp1_16_shard.csv")
```

```python
df = pd.DataFrame({"jolt":jolt["prover time (ms)"], 
                   "risczero":risczero["prover time (ms)"], 
                   "sp1_1":sp1_1["prover time (ms)"], 
                   "sp1_4":sp1_4["prover time (ms)"], 
                   "sp1_16":sp1_16["prover time (ms)"], 
                   "iters":jolt["iters"]})
df = df.drop([4])
ax= df.plot.barh(x="iters")
for container in ax.containers:
    ax.bar_label(container, labels=[f'{e:.2f}x' for e in container.datavalues / df["jolt"]])
plt.xlabel('prover time (ms)')
plt.ylabel('number of iterations')
plt.title('Prover time for sha2_chain in different zkVMs')
plt.savefig("sha2_chain_time.png")
```

```python
df = pd.DataFrame({"jolt":jolt["proof size (bytes)"],
                   "risczero":risczero["proof size (bytes)"], 
                   "sp1_1":sp1_1["proof size (bytes)"], 
                   "sp1_4":sp1_4["proof size (bytes)"], 
                   "sp1_16":sp1_16["proof size (bytes)"], 
                   "iters":jolt["iters"]})
df = df.drop([4])
ax= df.plot.barh(x="iters")
for container in ax.containers:
    ax.bar_label(container, labels=[f'{e:.2f}x' for e in container.datavalues / df["jolt"]])
plt.xlabel('proof size (bytes)')
plt.ylabel('number of iterations')
plt.title('Proof size for sha2_chain in different zkVMs')
plt.savefig("sha2_chain_size.png")
```
